## Notes
- Each time a number is found to the right of its successor, that incurs another pass. 

```python
def do_pass(val, arr):
    for i, item in enumerate(arr):
        print(f"Looking for {val} at {i}")
        if item == val:
            print(f"Found {val} at {i}")
            val += 1
        if val > max(arr):
            break
    print("pass completed.")
    return val
```

```python
def brute_force(arr):
    passes = 0
    val = min(arr)
    while val <= max(arr):
        if val not in arr:
            val += 1
        val = do_pass(val, arr)
        passes +=1
    return passes

```

```python
def solve(arr):
    places = {}
    for i, val in enumerate(arr):
        places[val] = i
    passes = 1
    keys = sorted(places.keys())
    i = 0
    j = 1
    while j < len(keys):
        val1 = keys[i]
        val2 = keys[j]
        if places[val1] > places[val2]:
            passes +=1
        i = j
        j += 1
    return passes
```

```python
import random

arr = list(set([random.randint(1, 100) for i in range(100)]))
print(arr)
print(brute_force(arr))
print(solve(arr))
```
